```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('./train.csv', parse_dates = True, date_parser = 'Date')
```

```python
train.head(20)
```

```python
train['Date'] = pd.to_datetime(train['Date'])
train['day_of_week'] = train['Date'].dt.dayofweek
```

```python
train.head()
```

```python
train = train.drop('Date', axis = 1)
train = train.drop('ID', axis = 1)
train.isnull()
```

```python
for c in train.columns:
    mean_value = train[c].mean()
    train[c] = train[c].fillna(mean_value)
```

```python
from sklearn.linear_model import Lasso
y_train = train['Footfall']
X_train = train.drop('Footfall', axis = 1)
```

```python
from sklearn.linear_model import LinearRegression
linreg = Lasso(alpha = 1e-5,normalize = True,max_iter = 1e5)
linreg.fit(X_train, y_train)
```

```python
y_train_pred = linreg.predict(X_train)
```

```python
import math
rss = math.sqrt(sum((y_train_pred - y_train)**2))
print(rss)
```
